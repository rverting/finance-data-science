```python
def fibonaci(n):
    suiteprec2=1
    suiteprec=1
    suite = suiteprec+suiteprec2
    while suite < n:
        suiteprec2=suiteprec
        suiteprec=suite
        suite=suiteprec2+suiteprec
    return suite

fibonaci(50)
```

```python
#Minimize
from scipy import optimize
x = np.linspace(0,2,100)
y = 1/3*x**3 - 3/5*x**2+2+np.random.randn(x.shape[0])/20
plt.scatter(x,y)
def f(x, a, b ,c ,d):
    return a * x**3 + b * x**2 + c * x + d
optimize.curve_fit(f, x, y) # paramètres et tableau de covariance
param, paramcov = optimize.curve_fit(f, x, y)
plt.plot(x, f(x, param[0], param[1], param[2], param[3]), c='red')
```

```python
def f(x):
    return x**2+15*np.sin(x)
x = np.linspace(-10,10,100)
plt.plot(x, f(x))
optimize.minimize(f, x0=-8) #minimul local
startpoint = -5
optimize.minimize(f, x0=startpoint) #changement de point de départ pour trouver le minimum global et non local
result = optimize.minimize(f, x0=-5).x
plt.scatter(result, f(result), s=100, c='r') #résultat de l'optim
plt.scatter(startpoint, f(startpoint), marker='+', s=100, c='black')
```

# Traitement de signal (transformation de Fourier)

```python
from scipy import signal #éliminer toute tendance linéaire que l'on a dans un signal.  
x= np.linspace(0,20,100)
y = x + 4*np.sin(x)+ np.random.randn(x.shape[0])
new_y = signal.detrend(y)
plt.plot(x,y)
plt.plot(x, new_y)
```

```python
#Transformation de Fourier : analyser les séquences qui sont présentes dans un signal périodique
x= np.linspace(0,30,1000)
y = 3*np.sin(x)+ 2*np.sin(x*5) + np.sin(10*x)
plt.plot(x,y)
```

```python
from scipy import fftpack
fourier = fftpack.fft(y)
power = np.abs(fourier)
frequences = fftpack.fftfreq(y.size)
plt.plot(np.abs(frequences), power)
```

# Transformation de Fourier sur signal brouillé

```python
x = np.linspace(0,30,1000)
y = 3*np.sin(x)+ 2*np.sin(x*5) + np.sin(10*x) + np.random.random(x.shape[0])*10
plt.plot(x,y)
```

```python
fourier = fftpack.fft(y)
power = np.abs(fourier)
frequences = fftpack.fftfreq(y.size)
plt.plot(np.abs(frequences), power)
```

```python
fourier[power<400] = 0 #boolean indexing pour nettoyer le bruit
```

```python
plt.plot(np.abs(frequences), np.abs(fourier))
```

```python
filtered_signal = fftpack.ifft(fourier)
plt.figure(figsize=(15,10))
plt.plot(x,y)
plt.plot(x,filtered_signal, c='orange', lw=4)
plt.show()
```

```python
#Stratégie de la tortue - Bitcoin
datatemp = prices.copy()
datatemp['buy'] = np.zeros(len(datatemp))
datatemp['sell'] = np.zeros(len(datatemp))
datatemp['RollingMax'] = datatemp['Close'].shift(1).rolling(window=28).max()
datatemp['RollingMin'] = datatemp['Close'].shift(1).rolling(window=28).min()
datatemp.loc[datatemp['RollingMax'] < datatemp['Close'], 'buy'] =1
datatemp.loc[datatemp['RollingMin'] > datatemp['Close'], 'sell'] = -1
```

```python
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
import yfinance as yf
import numpy as np
import pandas as pd

bitcoin = yf.Ticker("BTC-EUR")
prices = bitcoin.history(period="max") # attention ici Close = Adjusted Close
prices.head()
#le read csv peut s'écrire comme ceci pour faire du DateIndex 
# pd.reac_csv('BTC-EUR.csv', index_col='Date', parse_dates=True)

def strategybs(start, end, rolling):
    datatemp = prices[start:end].copy()
    datatemp['buy'] = np.zeros(len(datatemp))
    datatemp['sell'] = np.zeros(len(datatemp))
    datatemp['RollingMax'] = datatemp['Close'].shift(1).rolling(window=rolling).max()
    datatemp['RollingMin'] = datatemp['Close'].shift(1).rolling(window=rolling).min()
    datatemp.loc[datatemp['RollingMax'] < datatemp['Close'], 'buy'] =1
    datatemp.loc[datatemp['RollingMin'] > datatemp['Close'], 'sell'] = -1
    fig, ax = plt.subplots(2, figsize=(14,10), sharex=True)
    ax[0].plot(datatemp['Close'])
    ax[0].plot(datatemp['RollingMax'])
    ax[0].plot(datatemp['RollingMin'])
    ax[0].legend(['close', 'max', 'min'])
    ax[1].plot(datatemp['buy'], c='g')
    ax[1].plot(datatemp['sell'], c='r')
    ax[1].legend(['buy', 'sell'])
    
strategybs('2016','2019', 45)
```

```python
plt.figure(figsize=(12,10))
prices.loc['2019', 'Close'].plot(c="pink")
prices.loc['2019', 'Close'].resample('M').mean().plot(label='moyenne par mois', c='red', alpha=0.8, lw=4, ls='--')
prices.loc['2019', 'Close'].resample('W').mean().plot(label='moyenne par semaine', c='black', alpha=0.8, lw=3, ls=':')
plt.legend()
plt.show()
```

Range Min Max & Resample Function

MatplotLib & Pandas

```python
m = prices['Close'].resample('W').agg(['mean', 'std', 'min', 'max'])
#prices.loc['2019', 'Close'].resample('W').agg(['mean', 'std', 'min', 'max'])

plt.figure(figsize=(12,8))
m['2019']['mean'].plot(label='Moyenne hebdomadaire',c='r')
plt.fill_between(m.index, m['min'], m['max'], alpha=0.5, label="min-max par semaine")
plt.legend()
plt.show()
```

Moving Average

```python
plt.figure(figsize=(12,12))
prices.loc['2019','Close'].plot(label='Close', color='red', lw=3, ls=':')
prices.loc['2019','Close'].rolling(window=7, center=True).mean().plot(label='Moving average', color='blue', lw=3)
prices.loc['2019','Close'].rolling(window=14, center=True).mean().plot(label='Moving average', color='green', lw=3)
plt.legend()
plt.show()
```

Exponential moving Average

```python
plt.figure(figsize=(12,8))
prices.loc['2019-09','Close'].plot()
for i in np.arange(0.2,1,0.2):
    prices.loc['2019-09','Close'].ewm(alpha=i).mean().plot(label =f'alpha = {i}', ls="--", alpha=0.8)
plt.legend()
plt.show()
```
